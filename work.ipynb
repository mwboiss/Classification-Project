#### statistical testing

```python
#$H_0$: Those who churn pay the same average monthly charges as the whole population. 

#$H_α$ Those who churn pay a higher average monthly charge than the whole population.

churn_yes = train[train['churn'] == 1].monthly_charges
α = 0.05
population_mean = telco.monthly_charges.mean()

t , p = stats.ttest_1samp(churn_yes, population_mean)

if p/2 > α:
    print('Fail to reject Null')
elif t < 0:
    print('Fail to reject Null')
else:
    print('Reject Null')
    
print(f'\nt = {t}, \np = {p}')
```

Those who churn pay a higher average monthly charge.

## Statisitcal test

$H_0$: Those who churn had the same amount or less subscriptions then those that did not.

$H_α$: Those who churn had more subscriptions than those that did not. 

```python
churn = train[train.churn == 1]
no_churn = train[train.churn == 0]
churn.service_count.var(), no_churn.service_count.var()
```

```python
α = 0.05
t, p = stats.mannwhitneyu(churn.service_count, no_churn.service_count, alternative='greater')

if p < α:
    print('Reject null')
else:
    print('Fail to Reject null')
    
print(f'\nt = {t}, p = {p}')
```

According to the statistical test those that did churn had equal or less extra subscriptions.

## Statistical Test

$H_0$: Churn is independent of having a partner or dependents.

$H_α$: There is a relationship between churn and having a partner or dependents. 

```python
α = 0.05

partner_observed = pd.crosstab(train.churn, train.partner)
dependents_observed = pd.crosstab(train.churn, train.dependents)

pchi2 , pp, pdegf, pexpected = stats.chi2_contingency(partner_observed)
dchi2 , dp, ddegf, dexpected = stats.chi2_contingency(dependents_observed)

print(f'Partner P: {pp}, Dependents P: {dp}')

if pp < α:
    print('Reject null for partner')
else:
    print('Fail to reject null for partner')
    
if dp < α:
    print('Reject null for dependents')
else:
    print('Fail to reject null for dependents')
```

There is a relationship between churn and having a partner or dependents.
